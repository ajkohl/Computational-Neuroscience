```python
# Make a function for loading in the labels
def load_vdc_stim_labels(vdc_data_dir, sub):
    '''
    Load in the vdc stimulus labels
    
    Parameters:
        vdc_data_dir (string): The directory where the vdc data lives.
        subject_id (string): The subject ID whose stimulus labels we're loading

    Returns:
        stim_label_allruns (array): Stimulus labels concatenated for all runs.
    '''
    
    stim_label_allruns = []
    for run in range(1, vdc_n_runs + 1):
        in_file = os.path.join(vdc_data_dir, sub ,'ses-day2','design_matrix',f'{sub}_localizer_{run:02d}.mat')
        # Load in data from matlab
        stim_label = scipy.io.loadmat(in_file)
        stim_label = np.array(stim_label['data'])

        # Store the data
        if run == 1:
            stim_label_allruns = stim_label
        else:       
            stim_label_allruns = np.hstack((stim_label_allruns, stim_label))
    print("Loaded labels for", sub)
    return stim_label_allruns

# Note: you won't see the printed "Loaded labels for..." until you run the function with inputs (see next cell).
```

We now want to call this function and get its outputs.

```python
# Load the labels for this data. 
stim_label_allruns = load_vdc_stim_labels(vdc_data_dir, sub)

# Check to make sure they're the expected dimensions:
print(f'stim_label_allruns is of shape: {stim_label_allruns.shape}')
```

We need category labels to train a classifier for every timepoint in the fMRI data. These timepoints are also referred to as TRs.

This subject has 310 TRs but all others have 311 (no good reason, just a quirk of the dataset). Some TRs will not have a label since there was no stimulus presented (e.g., the fixation periods between blocks). 

Recall that the third row of the timing file is the start time of a trial in seconds since the start of the run. We can thus convert each time stamp to a specific TR by taking that time stamp and dividing by the TR duration (1.5 s). 

```python
# Preset variables
_, events = stim_label_allruns.shape
# How many events are there per run?
events_run = int(events / vdc_n_runs)

# Pre-allocate the array. 
# Since this is a small array, pre-allocation won't speed up the analysis notably,
# but this is useful for creating larger arrays!
stim_label_TR = np.zeros(vdc_TRs_run * 3)

# Loop through each run.
for run in range(vdc_n_runs):

    # Loop through each event in the run.
    for i in range(events_run):

        # What index in the concatenated label file are we looking for?
        time_idx = run * events_run + i

        # What is the time stamp at this index?
        time = stim_label_allruns[2, time_idx]

        # What TR does this timepoint refer to?
        TR_idx = int(time / vdc_TR) + (run * (vdc_TRs_run - 1))

        # Record the stimulus label at this TR.
        stim_label_TR[TR_idx] = stim_label_allruns[0, time_idx]
        
print(stim_label_TR)
```

```python
def label2TR(stim_labels, num_runs, TR, TRs_run):
    # Preset variables
    _, events = stim_labels.shape
    # How many events are there per run?
    events_run = int(events / num_runs)

    # Pre-allocate the array. 
    # Since this is a small array, pre-allocation won't speed up the analysis notably,
    # but this is useful for creating larger arrays!
    stim_labels_TR = np.zeros(TRs_run * 3)

    # Loop through each run.
    for run in range(num_runs):

        # Loop through each event in the run.
        for i in range(events_run):

            # What index in the concatenated label file are we looking for?
            time_idx = run * events_run + i

            # What is the time stamp at this index?
            time = stim_labels[2, time_idx]

            # What TR does this timepoint refer to?
            TR_idx = int(time / TR) + (run * (TRs_run - 1))

            # Record the stimulus label at this TR.
            stim_labels_TR[TR_idx] = stim_labels[0, time_idx]
    '''
    Match the stimulus labels for each TR.
    
    Parameters:
    ----------
    
        stim_labels : np.array
            Array containing the MATLAB stimulus labels for all the runs.
        
        num_runs : int 
            Number of runs that will be combined.
            
        TR : float
            Length of a TR in seconds.
            
        TRs_run : int
            Number of TRs per run.

    Returns:
    -------
    
        stim_labels_TR : np.array 
            Array containing the stimulus label for each TR.
    '''
    

    return stim_labels_TR
```

**1.3.2 Time shift the VDC labels**  
You will note that it takes approximately 4-6s for the BOLD response to peak after an event. Hence the brain's activity does not reflect what is occuring in the world currently, but rather what occurred in the recent past. To account for the hemodynamic lag, we can shift the timecourse of labels such that they correspond in time with the future brain response they produce. First let's plot this timecourse.

```python
n_conditions = len(vdc_label_dict)
palette = sns.color_palette('colorblind', n_colors=n_conditions)

# Create a sequence of timepoints, in seconds, that correspond with a TR.
tr_time = np.arange(0, (vdc_TRs_run - 1) * 1.5 + 1, 1.5)
# Get the time (seconds) that corresponds with stimulus labels.
time_vals = stim_label_allruns[2, 0:150]
# Get the stimulus labels at those timepoints.
labels = stim_label_allruns[0, 0:150]

f, ax = plt.subplots(1,1, figsize = (14, 5))
    
# Plot the label for each condition
for condition_index in range(n_conditions): 
    # Get the label name (face, scene, or object) that matches with a condition index.
    label = list(vdc_label_dict.keys())[condition_index] 
    # Create a mask for these labels only.
    temp_mask = label == labels
    # Scatter points related to this stimulus.
    ax.scatter(time_vals[temp_mask], labels[temp_mask], 
               color = palette[condition_index], marker = 'o')
ax.legend(vdc_label_dict.values())
    
# Plot the stimuli as a line. 
ax.plot(tr_time, stim_label_TR[0:vdc_TRs_run], c='orange', alpha = .5)

ax.set_yticks(list(vdc_label_dict.keys()))
ax.set_yticklabels(vdc_label_dict.values())

ax.set_title('Stimulus Presentation for Run 1')
ax.set_xlabel('Time (seconds)')
```

We need to incorporate this time-shift when we extract the BOLD signal for classification (so that the labels apply to the correct brain images). One way to accomplish this is to shift the labels and extract the BOLD signal for the non-zero labels. Given that one TR = 1.5 s and the typical hemodynamic lag is 4-5 s, we will shift the labels forwards by 3 TRs.

```python
# Shift the data a certain amount

# What is the lag (in seconds) between a stimulus onset and the peak bold response
print(f'Amount of time shift: {vdc_hrf_lag} s') 
shift_size = int(vdc_hrf_lag / vdc_TR)  # Convert the shift into TRs

# Create a function to shift the size
def shift_timing(label_TR, TR_shift_size):
    
    # Create a short vector of extra zeros
    zero_shift = np.zeros(TR_shift_size)

    # Zero pad the column from the top
    label_TR_shifted = np.hstack((zero_shift, label_TR))

    # Don't include the last rows that have been shifted out of the time line
    label_TR_shifted = label_TR_shifted[0:label_TR.shape[0]]
    
    return label_TR_shifted

# Apply the function
stim_label_TR_shifted = shift_timing(stim_label_TR, shift_size)


```

```python
# Add the boxcar to the plot

f, ax = plt.subplots(1,1, figsize = (12,5))
ax.plot(tr_time, stim_label_TR[0:vdc_TRs_run], c='blue',alpha=0.2)
ax.plot(tr_time, stim_label_TR_shifted[0:vdc_TRs_run], c='orange')

ax.set_yticks(list(vdc_label_dict.keys()))
ax.set_yticklabels(vdc_label_dict.values())

ax.set_ylabel('Stimulus labels')
ax.set_xlabel('Time (seconds)')

plt.legend(['Original', 'Shifted'])
```

## 2. Classification<a id="classification"></a>

We will now build a very basic classifier to categorize our data. 

In the example of email, spam mail has certain common features: sent from an unknown sender; sells products; asks for money deposits. These features are converted into mathematical vectors in a language feature space and computer programs are trained on these feature vectors,  _from known examples of relevant and spam emails_, to categorize emails as either relevant or spam. 

For brain activity measured by fMRI, the voxel signals serve as the features  that make up the brain's response to a stimulus. If we were showing pictures of faces, scenes, and objects, and collecting fMRI data while the subject was viewing the pictures, we know what picture was shown at each timepoint. The stimulus type at each timepoint serves as the stimulus label - just as the "label" of spam/not spam for emails. The voxel signals at each timepoint are features corresponding to the brain's response to the picture. Since we know both the voxel signals and the stimulus that was presented at a given timepoint, we can train a classifier to distinguish between pictures of faces, scenes, and objects.

Once the classifier training is accomplished, we  do not know how this classifier will perform when we give it a new set of voxel signals corresponding to a stimulus the classifier didn't see during training. To determine if the classifier can _generalize_ to accurately predict an stimulus label not presented during training, we test it on an _untrained_ dataset and determine its prediction accuracy. If the classifier chooses the right stimulus label above chance (random guessing will lead to a 33.33% accuracy for 3 categories), the classifier has read out from the brain the category of the stimulus.

We will try a classifier called a [support vector machine](https://en.wikipedia.org/wiki/Support-vector_machine) in this exercise.

### 2.1 Reshape data <a id="reshape"></a>

First, we need to extract the BOLD signal at the timepoints for which we have stimulus labels (faces, places, and objects). We will exclude the timepoints that correspond with fixation periods. 

```python
# Extract BOLD data for non-zero labels
def reshape_data(label_TR_shifted, masked_data_all):
    # Find the TRs that correspond with a stimulus label (not fixation)
    label_index = np.nonzero(label_TR_shifted)
    label_index = np.squeeze(label_index)
    # Pull out the voxel responses at those indices
    indexed_data = np.transpose(masked_data_all[:,label_index])
    nonzero_labels = label_TR_shifted[label_index] 
    return indexed_data, nonzero_labels

# Pull out the data from the PPA ROI for these time points
roi_name = "PPA"
roi_masked_data = epi_mask_data_all[vdc_all_ROIs.index(roi_name)]
bold_data, labels = reshape_data(stim_label_TR_shifted, roi_masked_data)

# What is the dimensionality of the data?g
print(f'The {roi_name} BOLD data is of dimensionality: {bold_data.shape[0]} TRs by {bold_data.shape[1]} voxels')
```

### 2.2 Leave-One-Run-Out (LORO) training and testing  <a id="loro_train_test"></a>

We have the vdc dataset to train our classifier. But where do we get an _unseen_ dataset to test the classifier's accuracy? To do this, we can use Leave-One-Run-Out (LORO) cross-validation. A typical fMRI experiment has multiple runs, wherein the same stimulus categories are shown in all runs. We can create an "unseen dataset" by setting one run of data aside when we train our classifier. This left-out run is now our unseen dataset -- the classifier was never trained on these data, so we can use it to test the accuracy of the classifier. 

Furthermore, we can loop through the runs, leaving one run out each time (also known as folding), and test multiple times. Let's say we have 3 runs: 1,2,3. Then, we can establish the sequence of training and testing, giving us 3 independent tests of the classifier accuracy, as shown in the table below. This process of splitting the available data to train and test classifiers is called cross-validation. You can also find more help on this topic [here](https://scikit-learn.org/stable/modules/cross_validation.html).

| Fold | Train Runs | Test Run|
| --- | --- | --- |
|1 |  1, 2 | 3 |
|2 |  2, 3 | 1 |
|3 |  1, 3 | 2 |
  
There are other ways you could do this that we will cover in the course. Sometimes it is an even better approach to leave a participant out in the analysis, so that way you can test whether your classifier transfers across people. Other times it is not appropriate to do your training and testing across runs. For instance, imagine you are studying how people learn two categories of information over an fMRI session: in this case, the runs are not interchangeable since in the early runs they won't have learned much compared to the later runs. In such cases we need to *carefully* split our data within a run into train and test sets.

#### 2.2.1. Manually create a left-out run.

**2.2.2 Normalization**

Normalization is an important step in classification. Normalization puts feature values in a similar range, reduces the variance across features, and prevents classifiers from being biased towards features with large values.


```python
def normalize(bold_data, run_ids):
    """Normalize BOLD data within each run
    
    Parameters:
    ----------
    bold_data: np.array [n_stimuli , n_voxels]
    run_ids: np.array or a list [n_stimuli,]
    
    Return:
    ------
    normalized_data: np.array, [n_stimuli , n_voxels]
    """
    
    scaler = StandardScaler()
    data = []
    for r in range(vdc_n_runs):
        data.append(scaler.fit_transform(bold_data[run_ids == r, :]))
    normalized_data = np.vstack(data)
    return normalized_data
```

#### 2.3.2. Build a classifier by training and testing on multiple folds.

We use scikit learn libraries to automate the task of creating multiple folds. The `PredefinedSplit` method does this automatically.

There are <a href="https://scikit-learn.org/stable/modules/classes.html#module-sklearn.model_selection">multiple methods</a> to splitting training and test data. No matter which method is adopted, you should always make sure that the train and test datasets are independent. As an example, if you do not separate the training and test dataset, you will get inflated accuracies (this is like taking a quiz after you have already been shown the answers). We will cover more of this aspect in future notebooks. Here's an example of re-using the same data for training and test and how it (artificially) inflates performance accuracy. ***NEVER*** do this.

```python
"""double dipping"""

# Create the classifier model.
model = LinearSVC(C=1) 

# Get training bold data and labels
X_train, y_train = bold_data_normalized[train_runs], labels[train_runs]

# Fit a SVM on all the data and all the labels.
model.fit(X_train, y_train)

# Score the model on the training data -- the same timepoints used in fitting the model. 
score = model.score(X_train, y_train)
print(f'Accuracy = {score}')

# If your results look too good, they probably are!!!!
```

Now we show how to split the dataset into training and testing in the correct way.

sklearn can handle the train-test split with [PredefinedSplit](http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.PredefinedSplit.html). This approach is particularly useful for LORO or when leaving a participant out.

```python
# Will store the scores for each run.
scores = []

# Split data according to run ids.
ps = PredefinedSplit(run_ids)

# Build the classifier. 
model = LinearSVC(C=1)

for train_index, test_index in ps.split():
    # Split the data according to the predefined split.
    X_train, X_test = bold_data_normalized[train_index], bold_data_normalized[test_index]
    y_train, y_test = labels[train_index], labels[test_index]
    
    # Fit the SVM.
    model.fit(X_train, y_train)
    
    # Calculate the accuracy on the held out run.
    score = model.score(X_test, y_test)
    scores.append(score)
    
print('Accuracy across folds:', scores)
```

**2.3.3 Classifier Performance**  
To put the accuracy of a classifier in context, we can compare it to an estimation of choosing the right answer by chance -- random guessing. In the VDC dataset, we have 3 equally represented sets of stimuli, so random guessing will result in an accuracy of 33.3333%. 

### 2.4 Perform the analysis on multiple  participants <a id="across_ppts"></a>

The next step is to run a classifier on a group of subjects. First we create a function that includes the codes the code above for classfiying one subject.

**Exercise 2:**<a id="ex2"></a> Convert the classifier code above into a function called `decode` with the inputs of data, labels, and run_labels.

```python
# Put code here. The function should output models and scores
def decode(data, labels,run_labels):
    """ Classify stimulus categories from BOLD data with LORO cross-validation.
    
    Parameters:
    ----------
    data : np.array, (n_stimuli x n_voxels)
        BOLD data for all runs, normalized within run, masked to exclude fixation timepoints.
        
    labels : np.array or list, (n_stimuli, )    
        What stimulus category was shown at this timepoint?
        
    run_labels : np.array or list, (n_stimuli, )
        What run does this timepoint belong to? 
    
    Return:
    ------
    models : list of LinearSVC models, (n_runs, )    
        Classification models fit at each loop of training/testing.
        
    scores : list of floats, (n_runs, )
        Classification accuracies for every fold.
    """
    
    # Will store the scores for each run.
    models = []
    scores = []

    # Split data according to run ids.
    ps = PredefinedSplit(run_labels)

    # Build the classifier. 
    model = LinearSVC(C=1)

    for train_index, test_index in ps.split():
        # Split the data according to the predefined split.
        X_train, X_test = data[train_index], data[test_index]
        y_train, y_test = labels[train_index], labels[test_index]

        # Fit the SVM.
        modelo = model.fit(X_train, y_train)
        models.append(modelo)

        # Calculate the accuracy on the held out run.
        score = model.score(X_test, y_test)
        scores.append(score)
    
    return models, scores
```

## 3. Modular vs. Distributed Processing <a id="mod"></a>

The discovery of the Fusiform Face Area (FFA; Kanwisher, McDermott, and Chun, 1997; see also McCarthy, Puce, Gore, & Allison, 1997) was ground breaking for the field. These studies found that the mean activity signal from the FFA for face stimuli was larger than that of houses or objects. They concluded that the FFA preferentially activates to faces, and by extension, it was inferred that all other cognitive processing must also be localized in brain regions yet to be discovered. This led to the **modular** view of face processing.

The local nature of processing was challenged by another study (Haxby et al., 2001). Instead of looking at the mean activity of a set of voxels, this study examined the _pattern of activity_ of a set of voxels. Thus, if the activity across a set of voxels is similar _on average_ for two conditions but the _pattern_ across the set of voxels for those two conditions differs, we can discriminate between the two conditions. Using this technique, it was shown that faces are not represented _solely_ in the FFA alone, but are distributed across a variety of brain regions. This led to the **distributed** view of face processing.

In this section, you will perform a decoding analysis in the FFA and the parahippocampal place area (PPA) using the VDC dataset. To recap, the FFA was shown as a face processing region and the PPA as a scene processing region. Specifically, you will analyze the patterns of activity in these ROIs in the following ways: 

>1. Can we discriminate scenes vs. objects in the FFA?  
      The FFA was shown to be a preferred region for face processing. If we can decode scenes vs. objects in this region, it implies that there is discriminable information for these two categories in the FFA. Thus, the FFA would not _just_ represent faces, but scenes and/or objects too. This would also mean that scenes are not exclusively represented in the PPA. 
           
>2. Can we discriminate faces vs. objects in the PPA?  
      The PPA was shown to be a preferred region for scene processing. If we can decode faces vs. objects in this region, it implies that there is discriminable information for these two categories in the PPA. Thus, the PPA would not _just_ represent scenes, but faces and/or objects too. This would also mean that faces are not exclusively represented in the FFA. 
      

### 3.1. Prepare the data for loading <a id="prep"></a>

The following provides an example for how to load data from both ROIs of one participant. 

```python
sub = 'sub-01'

for roi_name in ['FFA','PPA']:
  
    # Load the fMRI data
    epi_mask_data_roi = load_vdc_masked_data(vdc_data_dir, sub, [roi_name])

    # Load subject labels
    stim_label_allruns = load_vdc_stim_labels(vdc_data_dir, sub) 

    # load run ids 
    run_ids = stim_label_allruns[5,:] - 1 

    # Convert the timing into TR indexes
    TRs_run = int(epi_mask_data_roi[0].shape[1] / vdc_n_runs)
    stim_label_TR = label2TR(stim_label_allruns, vdc_n_runs, vdc_TR, TRs_run)

    # Shift the data some amount
    stim_label_TR_shifted = shift_timing(stim_label_TR, shift_size)

    # Select and reshape ROI data 
    bold_data_ROI, labels = reshape_data(
      stim_label_TR_shifted, epi_mask_data_roi[0])

    # normalize ROI data
    bold_ROI_normalized = normalize(bold_data_ROI, run_ids)
```

**Exercise 4:**<a id="ex4"></a> Encapsulate the above code in a function that returns the normalized BOLD activity of a single participant for a single ROI, as well as the labels and run_ids:

`bold_ROI_normalized, labels, run_ids = load_classification_data(sub,roi)`

Use this function for the next two classification exercises. 

```python
# Put code here
def load_classification_data(sub, roi):
    """ Get normalized BOLD activity, labels, and run_ids for a given subject and ROI.
    
    Parameters:
    ----------
    sub : string
        Subject id (sub-XX).
        
    roi : string    
        Which region are we loading data from? 
    
    
    Return:
    ------
    bold_ROI_normalized : np.array , (n_events, n_voxels)    
        Normalized (within run) BOLD data for a given ROI
        
    labels: np.array , (n_events, )
        Stimulus labels to correspond with the events in the BOLD data.
    
    run_ids: np.array, (n_events, )
        Run labels for each event.
    """
    # Load the fMRI data
    epi_mask_data_roi = load_vdc_masked_data(vdc_data_dir, sub, [roi])

    # Load subject labels
    stim_label_allruns = load_vdc_stim_labels(vdc_data_dir, sub) 

    # load run ids 
    run_ids = stim_label_allruns[5,:] - 1 

    # Convert the timing into TR indexes
    TRs_run = int(epi_mask_data_roi[0].shape[1] / vdc_n_runs)
    stim_label_TR = label2TR(stim_label_allruns, vdc_n_runs, vdc_TR, TRs_run)

    # Shift the data some amount
    stim_label_TR_shifted = shift_timing(stim_label_TR, shift_size)

    # Select and reshape ROI data 
    bold_data_ROI, labels = reshape_data(
      stim_label_TR_shifted, epi_mask_data_roi[0])
    
    
    # normalize ROI data
    bold_ROI_normalized = normalize(bold_data_ROI, run_ids)
    
    return bold_ROI_normalized, labels, run_ids
```
